### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import random as rnd

#Funtion to roll the dice.
def experimento(n):
    cont = 0
    for i in range(n):
        dado1 = rnd.randint(1,6)
        dado2 = rnd.randint(1,6)
        suma = dado1 + dado2
        if suma > 7 or suma%2 == 0: cont += 1
    return print('la probabilidad es: ' + str(cont/n))

experimento(1000)

# la probabilidad teórica es 2/3=0.66
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import random as rnd
import numpy as np

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'    
        
#Initialize important variables
n_simulations = 10000
exito_a = 0
exito_b = 0
exito_c = 0

for i in range(n_simulations):
    # array inicial para guardar cantidad de cada color en este orden: White, Red, Green 
    count_col = np.zeros(3)

    # elijo al azar 5 pelotas
    for i in range(5):
        sorteo = d[rnd.randint(0, len(d)-1)]
        if sorteo == 'White': count_col[0] += 1
        if sorteo == 'Red': count_col[1] += 1
        if sorteo == 'Green': count_col[2] += 1
    
    # acumulo éxitos en parte a (asumo que la condición es and)
    if np.array_equal(count_col, np.array([3,2,0])): exito_a += 1 # sirve solo un resultado

    # acumulo éxitos en parte b
    if count_col.max() == 5: exito_b += 1 # si un elemento es 5, los otros 2 son 0

    # acumulo éxitos en parte c (como no estoy segura, agrego parte a con condicion or)
    if count_col[0] == 3 or count_col[1] == 2: exito_c += 1 # sirven todos los casos con 3 blancas o 2 rojas

print('la probabilidad de sacar 3 blancas y 2 rojas es ' + str(round(100*exito_a/n_simulations,2)) + '%')    
print('la probabilidad de sacar las 5 del mismo color es ' + str(round(100*exito_b/n_simulations,2)) + '%')
print('la probabilidad de sacar 3 blancas o 2 rojas es ' + str(round(100*exito_c/n_simulations,2)) + '%')    

# probabilidad teórica de sacar 3 blancas y 2 rojas: ((1/6)**3)*((2/6)**2) = 0.0005144033
# probabilidad teórica de sacar todas del mismo color: (1/6)**5+(2/6)**5+(3/6)**5 = 0.03549383 
# probabilidad teórica de sacar 3 blancas o 2 rojas: (1/6)**3 + (2/6)**2 - ((1/6)**3)*((2/6)**2) = 0.1152263

### hay bastante variabilidad de un resultado a otro, al no fijar una semilla
### no sé si calculé mal la probabilidad teórica de la parte c, no se parece a lo hallado en las simulaciones
```
